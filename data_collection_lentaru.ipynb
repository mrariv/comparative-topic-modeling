# Collecting Corpus Data for Thesis

For the goal of our thesis we need a representative corpus of news documents

```python
import pandas as pd
from datasets import load_dataset
```

## 1. Collect Data Before 2019 (yutkin Kaggle repo, 1999-2019)

Source: [Kaggle Dataset](https://www.kaggle.com/datasets/yutkin/corpus-of-russian-news-articles-from-lenta)

```python
df_before_2019 = pd.read_csv("/Users/i.mrasov/Downloads/lenta-ru-news.csv")
```

```python
print('Dataset shape:', df_before_2019.shape)
print('Columns:', df_before_2019.columns)
df_before_2019.head()
```

We are working with large corpus so to ensure the domain-specific approach we would need to filter data based on the `topic` values

```python
print(df_before_2019['topic'].value_counts())
```

Initial choices:

- Экономика             79528

- Бизнес                 7399

Let's filter news articles according to these topics. This way we can ensure domain specificity of news and focus on socio-economic documents

```python
df_before_2019_economic = df_before_2019[(df_before_2019['topic'] == 'Экономика') | (df_before_2019['topic'] == 'Бизнес')]
print(df_before_2019_economic.shape)
```

```python
df_before_2019_economic.head()
```

Checking for data — deleting duplicates and empty rows

```python
df_before_2019_economic.isna().sum()
```

```python
print(df_before_2019_economic.duplicated().sum())
print(df_before_2019_economic.duplicated(subset=["url"]).sum())
```

Nice, no empty rows and duplicates. Let's convert data to correct dtypes

```python
df_before_2019_economic.dtypes
```

```python
df_before_2019_economic["date"] = pd.to_datetime(df_before_2019_economic["date"], errors="coerce")
```

```python
df_before_2019_economic.dtypes
```

## 2. Collect Data After 2019 (data-silence HuggingFace, 2019-2024)

Source: [HuggingFace](https://huggingface.co/datasets/data-silence/lenta.ru_2-extended)

```python
df_from_2019 = pd.read_csv("hf://datasets/data-silence/lenta.ru_2-extended/final_lenta_new.csv.gz")
```

```python
print(df_from_2019.shape)
print(df_from_2019.columns)
df_from_2019.head()
```

```python
print(df_from_2019['topic'].value_counts())
```

Here we choose once again these topics:

- Экономика             57548

Бизнес is not included, only has 37 articles (insignificant)

```python
df_from_2019_economic = df_from_2019[df_from_2019['topic'] == 'Экономика']
print(df_from_2019_economic.shape)
```

```python
df_from_2019_economic.head()
```

Checking for data — deleting duplicates and empty rows

```python
df_from_2019_economic.isna().sum()
```

```python
print(df_from_2019_economic.duplicated().sum())
print(df_from_2019_economic.duplicated(subset=["url"]).sum())
```

Once again, no empty rows and duplicates. Let's convert data to correct dtypes

```python
df_from_2019_economic.dtypes
```

```python
df_from_2019_economic["date"] = pd.to_datetime(df_from_2019_economic["date"], errors="coerce")
```

```python
df_from_2019_economic.dtypes
```

Also, let's delete `links` column as it is not needed for topic modeling

```python
df_from_2019_economic = df_from_2019_economic.drop(columns=['links'])
df_from_2019_economic.head()
```

And to concatenate the latter dataset with the former we need to reorder the columns 

```python
print('Structure before 2019:', df_before_2019_economic.columns)
print('Structure from 2019:', df_from_2019_economic.columns)
```

```python
df_from_2019_economic = df_from_2019_economic[['url', 'title', 'news', 'topic', 'tags', 'date']]
df_from_2019_economic = df_from_2019_economic.rename(columns={'news': 'text'})
df_from_2019_economic.head()
```

## 3. Concatenating Two Datasets 

```python
df_combined = pd.concat([df_before_2019_economic, df_from_2019_economic], ignore_index=True)
print('Full corpus shape:', df_combined.shape)
df_combined.head()
```

One important check for duplicates in a combined datasets 

```python
print(df_combined.duplicated(subset=["url"]).sum())
```

This is our full corpus for the work. Store it in .csv for now

```python
df_combined.to_csv("lenta_socioeconomic_corpus.csv", index=False)
```
